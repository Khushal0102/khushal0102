```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]
plt.grid()
plt.plot(year,price)
```

```python
x =pd.read_csv('/content/courses.csv')
x
plt.plot(x['course_id'],x['price'])
plt.plot(x['course_id'],x['discount'])
```

```python
plt.plot(x['course_id'],x['price'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Price')
plt.plot(x['course_id'],x['discount'],color='tab:brown',linestyle='dashdot',linewidth=3,marker='D',markersize=8,label='Discount')
plt.xlabel('Course_id')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```
